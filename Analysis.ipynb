```python
#Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Load datset
df=pd.read_csv('F:/OneDrive/Documents/OASIS_INFOBYTE_INTERNSHIP/Unemployment_Analysis/Data/Unemployment_Rate_upto_11_2020.csv')
df.head()
```

```python
#Basic analysis
df.info()
df.describe()
```

```python
#Data cleaning
df.isnull().sum()
```

```python
#Rename columns
df.columns=["State","Date","Frequency","Estimated Unemployment Rate","Estimated Employed","Estimated Labour Participation Rate (%)","Region","Longitude","Latitude"]
df.head()
```

```python
#Change of date format
df["Date"]=pd.to_datetime(df["Date"])
df.head()
```

```python
#Data visualization
#Unemployment Rate Over Time
plt.figure(figsize=(10,5))
sns.lineplot(x='Date', y='Estimated Unemployment Rate', data=df)
plt.title('Unemployment Rate Over Time')
plt.savefig("F:/OneDrive/Documents/OASIS_INFOBYTE_INTERNSHIP/Unemployment_Analysis/Graphs/Unemployment Rate Over Time.png",dpi=100, bbox_inches='tight')
plt.show()
```

```python
#Region wise analysis
plt.figure(figsize=(10,5))
sns.barplot(x='Region', y='Estimated Unemployment Rate', data=df)
plt.xticks(rotation=90)
plt.savefig("F:/OneDrive/Documents/OASIS_INFOBYTE_INTERNSHIP/Unemployment_Analysis/Graphs/Estimated Unemployment Rate.png",dpi=100, bbox_inches='tight')
plt.show()
```

```python
#COVID Impact
covid_data = df[df['Date'].dt.year == 2020]

plt.figure(figsize=(10,5))
sns.lineplot(x='Date', y='Estimated Unemployment Rate', data=covid_data)
plt.title('Unemployment During COVID (2020)')
plt.savefig("F:/OneDrive/Documents/OASIS_INFOBYTE_INTERNSHIP/Unemployment_Analysis/Graphs/Unemployment During COVID.png",dpi=100, bbox_inches='tight')
plt.show()
```

```python
#INSIGHTS 
#COVID-19 caused a sharp, short-term shock to employment with peak unemployment in May–June 2020
#The recovery was rapid but incomplete, transitioning from a V-shaped rebound to slower stabilization
#The impact varied significantly across regions, with North and East being most affected, while West remained relatively stable
#Increased variability during peak months indicates economic uncertainty and data instability
#The delayed peak highlights a lag effect between policy actions and labor market response
```
